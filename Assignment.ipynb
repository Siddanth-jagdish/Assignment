### Removal of punctuation, stopwords and lemmatizing the review text

```python
import string
```

```python
pun = string.punctuation
```

```python
pun
```

```python
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
stopwords = list(STOP_WORDS)
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def clean_it(sent):
    file = nlp(sent)
    
    list1 = []
    for elem in file:
        if elem.lemma_ != "-PRON-":
            i = elem.lemma_.lower().strip()
        else:
            i = elem.lower_
        list1.append(i)
    tokens = []
    for elem in list1:
        if elem not in stopwords and elem not in pun:
            tokens.append(elem)
    return tokens
```

```python
clean_it("This kindle is light and easy to use especially at the biggest beach!!!")
```

### using Tfidf vectorizer to focus on the frequency of the words present in the corpus and also to know the importance of words.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(tokenizer = clean_it)
```

```python
from sklearn.model_selection import train_test_split
```

### storing the review text into variable x and the target variable into y

```python
x = df1["reviews.text"]
```

```python
y = df1["reviews.doRecommend"]
```

### Making use of train_test_split and spliting the data set for about .33 to test data before we fit in the model

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

### Importing Logistic Regression algorithm to make predictions and to check Accuracy

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
logreg = LogisticRegression(C=1e5)
```

```python
clf = Pipeline([('tfidf',tfidf),('model', logreg)])
```

```python
clf.fit(x_train,y_train)
```

```python
ypred = clf.predict(x_test)
```

### Performance Evaluation using Confusion matrix, classification reports and Accuracy score

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test, ypred))
print(confusion_matrix(y_test, ypred))
accuracy_score(y_test, ypred)
```

### WordCloud Visualization

```python
from wordcloud import WordCloud
```

```python
df["sentiment"] = df["reviews.rating"].apply(lambda score: "positive" if score > 3 else "negative")
```

```python
df
```

```python
def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=stopwords,
        max_words=300,
        max_font_size=40, 
        scale=3,
        random_state=1 # chosen at random by flipping a coin; it was heads
    ).generate(str(data))
    
    fig = plt.figure(1, figsize=(8, 8))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.4)

    plt.imshow(wordcloud)
    plt.show()
    
show_wordcloud(df["reviews.title"])
```

### the negative reviews wordcloud

```python
show_wordcloud(df[df["reviews.rating"] == 1]["reviews.title"], title = "Low scored words")
```

### the positive reviews wordcloud

```python
show_wordcloud(df[df["reviews.rating"] == 5]["reviews.title"], title = "High scored words")
```
